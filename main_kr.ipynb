# **서론**

## **YZ 세대란 누구인가?**

'Y세대'는 지난 1997년에 미국에서 2000년, 즉 Y2000의 주역이 될 세대를 부르면서 생겨난 용어이다. '밀레니엄세대'라고도 불리며, 베이비붐세대가 낳았다고 해서 '에코 세대'라고도 합니다. 'Z세대'는 1990년대 중반에서 2000년대 초반에 걸쳐 출생한 젊은 세대를 이르는 말로, 밀레니엄 세대(Y세대)의 뒤를 잇는 인구 집단입니다. 즉, YZ 세대란 1980년대 초\~2000년대 초 출생한 Y세대와 1990년대 중반\~2000년대 초반 출생한 Z세대를 통칭하는 말입니다.

또한, 디지털 환경에 익숙한 YZ세대는 모바일을 우선적으로 사용하고, 최신 트렌드와 남과 다른 이색적인 경험을 추구하는 특징을 지니고 있으며,  SNS를 기반으로 유통시장에서 강력한 영향력을 발휘하는 소비 주체로 부상하고 있습니다. 이런 특징이 주식시장에도 반영이 되는지 살펴보겠습니다.

```
def create_icn_mdi(df_mergedata_kr):
  df_temp_merge = pd.DataFrame({'영업점단말':df_mergedata_kr[df_mergedata_kr['orr_mdi_dit_cd']== 1]['ivs_icn_cd'].value_counts()})
  df_temp_merge = df_temp_merge.reset_index()
  df_temp = pd.DataFrame({'유선단말':df_mergedata_kr[df_mergedata_kr['orr_mdi_dit_cd']== 2]['ivs_icn_cd'].value_counts()})
  df_temp = df_temp.reset_index()
  df_temp_merge = pd.merge(df_temp_merge,df_temp,how='outer')
  df_temp = pd.DataFrame({'무선단말MTS':df_mergedata_kr[df_mergedata_kr['orr_mdi_dit_cd']== 3]['ivs_icn_cd'].value_counts()})
  df_temp = df_temp.reset_index()
  df_temp_merge = pd.merge(df_temp_merge,df_temp,how='outer')
  df_temp = pd.DataFrame({'HTS':df_mergedata_kr[df_mergedata_kr['orr_mdi_dit_cd']== 4]['ivs_icn_cd'].value_counts()})
  df_temp = df_temp.reset_index()
  df_temp_merge = pd.merge(df_temp_merge,df_temp,how='outer')

  df_icn_merge = pd.melt(df_temp_merge, id_vars=['index'])
  df_icn_merge.rename(columns={'index':'고객투자성향','variable':'주문매체','value':'거래건수'},inplace=True)
  df_icn_merge.loc[(df_icn_merge['고객투자성향'] == '00'),'고객투자성향']='정보제공미동의'
  df_icn_merge.loc[(df_icn_merge['고객투자성향'] == '01'),'고객투자성향']='안정형'
  df_icn_merge.loc[(df_icn_merge['고객투자성향'] == '02'),'고객투자성향']='안정추구형'
  df_icn_merge.loc[(df_icn_merge['고객투자성향'] == '03'),'고객투자성향']='위험중립형'
  df_icn_merge.loc[(df_icn_merge['고객투자성향'] == '04'),'고객투자성향']='적극투자형'
  df_icn_merge.loc[(df_icn_merge['고객투자성향'] == '05'),'고객투자성향']='공격투자형'
  df_icn_merge.loc[(df_icn_merge['고객투자성향'] == '09'),'고객투자성향']='전문투자자형'
  df_icn_merge.loc[(df_icn_merge['고객투자성향'] == '-'),'고객투자성향']='해당 사항 없음'
  return df_icn_merge

def create_grd_mdi(df_mergedata_kr):
  df_temp_merge = pd.DataFrame({'영업점단말':df_mergedata_kr[df_mergedata_kr['orr_mdi_dit_cd']== 1]['tco_cus_grd_cd'].value_counts()})
  df_temp_merge = df_temp_merge.reset_index()
  df_temp = pd.DataFrame({'유선단말':df_mergedata_kr[df_mergedata_kr['orr_mdi_dit_cd']== 2]['tco_cus_grd_cd'].value_counts()})
  df_temp = df_temp.reset_index()
  df_temp_merge = pd.merge(df_temp_merge,df_temp,how='outer')
  df_temp = pd.DataFrame({'무선단말MTS':df_mergedata_kr[df_mergedata_kr['orr_mdi_dit_cd']== 3]['tco_cus_grd_cd'].value_counts()})
  df_temp = df_temp.reset_index()
  df_temp_merge = pd.merge(df_temp_merge,df_temp,how='outer')
  df_temp = pd.DataFrame({'HTS':df_mergedata_kr[df_mergedata_kr['orr_mdi_dit_cd']== 4]['tco_cus_grd_cd'].value_counts()})
  df_temp = df_temp.reset_index()
  df_temp_merge = pd.merge(df_temp_merge,df_temp,how='outer')

  df_grd_merge = pd.melt(df_temp_merge, id_vars=['index'])
  df_grd_merge.rename(columns={'index':'고객등급','variable':'주문매체','value':'거래건수'},inplace=True)
  df_grd_merge.loc[(df_grd_merge['고객등급'] == '01'),'고객등급']='탑클래스'
  df_grd_merge.loc[(df_grd_merge['고객등급'] == '02'),'고객등급']='골드'
  df_grd_merge.loc[(df_grd_merge['고객등급'] == '03'),'고객등급']='로얄'
  df_grd_merge.loc[(df_grd_merge['고객등급'] == '04'),'고객등급']='그린'
  df_grd_merge.loc[(df_grd_merge['고객등급'] == '05'),'고객등급']='블루'
  df_grd_merge.loc[(df_grd_merge['고객등급'] == '09'),'고객등급']='등급 없음'
  df_grd_merge.loc[(df_grd_merge['고객등급'] == '_ '),'고객등급']='해당 사항 없음'
  return df_grd_merge
```

## YZ 세대와 비 YZ 세대의 손익률
- 모든 거래를 비교해서 매수와 매도 데이터가 모두 있는 데이터를 사용했습니다.
- 배당금에 대한 자료는 없어서 무시하고 데이터를 분석했습니다.
- 수익률은 다음식을 사용해서 계산했습니다.$$(매도금 - 매수금) / 매수금$$
- 구매한 적 있는 종목을 재구매할 경우 평균 단가를 계산해서 수익률을 구했습니다.
- 평균단가는 다음식을 사용해서 계산했습니다.$$(평균단가 \times 보유수량 + 체결단가 \times 체결수량) / (보유수량 + 체결수량)$$
- 비율은 소수점 2자리까지 계산했습니다.
